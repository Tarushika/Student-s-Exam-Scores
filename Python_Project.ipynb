```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student Exam Score.csv")
print(df.head(5))
```

```python
df.isnull().sum()
```

# Drop Unnamed Column

```python
df = df.drop("Unnamed: 0",axis=1)
print(df.head())
```

# Gender Distribution

```python
plt.figure(figsize = (4,5))
ax = sns.countplot(data = df, x = "Gender")   #seaborn
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()    #matplotlib 
```

```python
# From the above Chart I have Analysed  that the:- 
# Number of Females in data is more then the number of Males.
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot= True)
plt.title("Relation between Parents's Education and Student's Score")
plt.show()
```

```python
# From the above Chart have Concluded that:- 
# The Education of the Parents have a good impact on their Scores.
```

```python
gb_1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb_1)
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gb_1, annot = True)
plt.title("Relation between Parent's Marital Status and Student's Score")
plt.show()
```

```python
# From the above Chart have Concluded that:- 
# That there is no / negligible impact on the Student's Score due to their Parent's Marital Status.
```

```python
sns.boxplot(data = df, x = "MathScore")
plt.title("Student's Maths Score")
plt.show()
```

```python
sns.boxplot(data = df, x = "ReadingScore")
plt.title("Student's Reading Score")
plt.show()
```

```python
sns.boxplot(data = df, x = "WritingScore")
plt.title("Student's Writing Score")
plt.show()
```

```python
# From the above 3 Chart have Concluded that:- 
# The Student's are comparatively weak in Maths instead of Reading and Writing
```

# Distribution of Ethnic Group 

```python
groupA = df.loc[(df['EthnicGroup']=="group A")].count()   # only present in Group A
groupB = df.loc[(df['EthnicGroup']=="group B")].count()   # only present in Group B
groupC = df.loc[(df['EthnicGroup']=="group C")].count()   # only present in Group C
groupD = df.loc[(df['EthnicGroup']=="group D")].count()   # only present in Group D
groupE = df.loc[(df['EthnicGroup']=="group E")].count()   # only present in Group E

l = ["group A","group B","group C","group D","group E"]
my_list = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"],groupE["EthnicGroup"]]

print(my_list)
plt.pie(my_list, labels = l , autopct ="%1.2f%%" )
plt.title("Distribution of Ethnic Groups")
plt.show()
```
